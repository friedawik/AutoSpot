## Peak detection using multiple contours levels

This tool uses matplotlib.pyplot.contour function to assemble mountain like features by grouping overlapping contours. The features are then filtered by a set of paramters that the user can tune.

Some important notes:

- The tools will ignore all open contours, thus all peaks who's lower contours are not fully included in a patch may be excluded. One way of getting around this problem, could be to include an overlapping buffer zone in the image patches to make sure all contours of reasonable size are included.

- About patches: The patches will have varying pixel value variation, which will impact the contour level resolution. Patches with high peak intensities will have greate pixel value difference between each level. This may lead to lower intensity peaks being more readily found in less contrasted images.

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
#import tifffile
import pandas as pd
from codes.tools import *
from codes.spot_detection_tool import SpotDetector
%matplotlib inline
```

```python
def load_data(img_id):
    image_path = '../../data/patch_256/test/images/' + img_id + '.tif'
    mask_path = '../../data/patch_256/test/masks/' + img_id + '_masks.png'
    img = cv2.imread(image_path,  cv2.IMREAD_UNCHANGED)
    img = img[:,:,1]
    masks = cv2.imread(mask_path,  cv2.IMREAD_UNCHANGED)
    masks = masks //255
    return img, masks
```

### Results from image patch
- Increasing the 'levels' parameter will result in more contours stacked on the top of each other, and therefore higher peak detection
- min_area: Excludes any contours smaller than the limit
- max_area: Excludes any contours greater than the limit
- min_z: Excludes any contours who's increase in pixel value between base and peak contour is smaller than the limit

```python
# get training image
start_x = 2
start_y = 2
#image_name = f'MF_MaxIP_3ch_2_000_230623_544_84_F_XY1_x{start_x}_y{start_y}'
#image_name = f'MF_MaxIP_3ch_2_000_230623_544_84_S4h_XY2_x{start_x}_y{start_y}'
img_id = f'MF_MaxIP_3ch_2_000_230623_544_84_R1h_XY5_x{start_x}_y{start_y}'
img, masks = load_data(img_id)
```

### Varying the 'level' parameter
- Increasing the number of stacked contours levels typically results in more peaks being found since the z difference will be calculated closer to the base and peak contours
- The difference is most notably when 'levels' is a small number

```python
min_area= 3
max_area= 200
levels = 15
min_z = 10
spots=SpotDetector(img,contour_levels=levels, min_area=min_area, max_area=max_area,min_circularity = 0.3, z_min_dist = min_z)
peak_dict = spots.peak_dict
plot_prediction(img, peak_dict,masks)
precision, recall, tp, fn, fp , masks_eval= evaluate_results(peak_dict, masks)
```

```python
min_area= 3
max_area= 200
levels = 150
min_z = 10
levels = np.arange(10, 150, 2)
test_z = np.arange(10, 80, 10)

num_peaks = [np.zeros(len(levels)) for _ in range(len(test_z))]
for i, level in enumerate(levels):
    spots=SpotDetector(img,contour_levels=level, min_area=min_area, max_area=max_area,min_circularity = 0.3, z_min_dist = min_z)
    peak_dict = spots.peak_dict
    for k, z_value in enumerate(test_z):
        filtered_dict = {key: value for key, value in peak_dict.items() if value['z_dist'] >= z_value}
        num_peaks[k][i] = len(filtered_dict)
    
for i,peaks in enumerate(num_peaks):
    plt.plot( levels,peaks,label=f'{test_z[i]}')


plt.legend(title='Min. z dist:')
plt.xlabel('Contour levels')
plt.ylabel('Detected peaks')
plt.grid()
plt.title(f'Pixel detection at varying level and z_dist')
plt.show()
```

### The tricky fed cells
As the cells engage in autophagy for energy provision, the autophagic vesicles grow and accumulate Atg8a. This typically results in lower background pixel values and higher peak values, which increases the contrast between the spots and the background, making spot classification easier. Therefore, it is not surprising that fed cells, which contain fewer and smaller autophagic vesicles, are harder to classify. The image below shows the results of the SpotDetector on an image patch of fed cells. While only two of the predicted peaks are inside a GT mask, several other spots that share similar characteristics such as size, intensity, and shape have been classified as 'false positives'. The masks that have not been detected, have a less 'spot-like' aspect that most other GT mask.  
- The SpotDetector finds the 6 most spot-like features
- Only two of them are segmented as a spot in the GT segmentation
- Three less bright and sharp areas have been segmented as GT mask

```python
img_id = 'MF_MaxIP_3ch_2_000_230623_544_84_F_XY4_x6_y5'
img, masks = load_data(img_id)
min_area= 3
max_area= 200
levels = 200
min_z = 100
spots=SpotDetector(img,contour_levels=levels, min_area=min_area, max_area=max_area,min_circularity = 0.3, z_min_dist = min_z)
peak_dict = spots.peak_dict
plot_prediction(img, peak_dict,masks)
precision, recall, tp, fn, fp , masks_eval= evaluate_results(peak_dict, masks)
```
